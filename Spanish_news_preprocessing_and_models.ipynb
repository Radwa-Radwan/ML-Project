**Real news (Web Scraping)**

```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
import io
df_spanish_web = pd.read_csv(io.BytesIO(uploaded['true_1.csv']))
df_spanish_web = pd.DataFrame(df_spanish_web)
df_spanish_web
```

**Merge Dataset**

```python
news_final= df_final_spanish.append(df_spanish_web, ignore_index = True)
news_final
```

**Set 1 and 0 for fake and real news**

```python
is_fake = ['fake']
news_final['is_fake'] = news_final['clase'].isin(is_fake).astype(int)
news_final
```

**Tokenize Text**

```python
import nltk
import gensim
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('stopwords')
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS



nltk.download('punkt');
def tokenize(column):
    tokens = nltk.word_tokenize(column)
    return [w for w in tokens if w.isalpha()]  
news_final['tokenized_text'] = news_final.apply(lambda x: tokenize(x['texto']), axis=1)
news_final.head()
```

**Stopwords**

```python
from nltk.corpus import stopwords


def remove_stopwords(column_2):
  x=[token for token in column_2 if token.lower() not in stopwords.words("spanish")]
  text = ' '.join(x)   
  return text

news_final['stopwords_removed_text'] = news_final.apply(lambda x: remove_stopwords(x['tokenized_text']), axis=1)

news_final
```

**Stemming**

```python
def apply_stemming(tokenized_column):
  stemmer = PorterStemmer() 
  w=[stemmer.stem(word).lower() for word in tokenized_column]
  text_1 = ''.join(w)  
  return text_1
news_final['porter_stemmed_text'] = news_final.apply(lambda x: apply_stemming(x['stopwords_removed_text']), axis=1)
news_final
```

**Train and test**

```python
from nltk.probability import FreqDist
from nltk.corpus import names  
import numpy as np
from string import punctuation
from PIL import Image 

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

X = news_final['porter_stemmed_text']
y = news_final['is_fake']



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, shuffle=True)

```

**Logistic Regression Model**

```python

# Creating model
bundled_pipeline = Pipeline([("tfidf", TfidfVectorizer()), ("lr", LogisticRegression())])


#Cross Validation
from sklearn.model_selection import train_test_split


import pandas as pd
from sklearn.model_selection import cross_validate
cv_results = cross_validate(bundled_pipeline, X, y, return_estimator=True)
pd.DataFrame(cv_results)
```

```python
from sklearn.model_selection import ShuffleSplit

cv = ShuffleSplit(n_splits=10, test_size=0.2)
cv_results = cross_validate(bundled_pipeline, X, y,
                            cv=cv, scoring="neg_mean_absolute_error",
                            return_train_score=True, n_jobs=2)
cv_results = pd.DataFrame(cv_results)
print (cv_results)

scores = pd.DataFrame()
scores[["train error", "test error"]] = -cv_results[
    ["train_score", "test_score"]]
    
import matplotlib.pyplot as plt
scores.plot.hist(bins=50, edgecolor="black")
plt.xlabel("Mean absolute error")
_ = plt.title("Train and test errors distribution via cross-validation")
```

```python
# Making prediction on test set
bundled_pipeline.fit(X_train, y_train)
y_pred = bundled_pipeline.predict(X_test)

print(classification_report(y_test, y_pred))
print('Accuracy:',accuracy_score(y_test, y_pred))
print('F1 score:',f1_score(y_test, y_pred))
print('ROC/AUC score:',roc_auc_score(y_test, y_pred))


from sklearn.metrics import mean_squared_error
import math
MSE=mean_squared_error(y_test, y_pred)
 
print("Root Mean Square Error:\n", MSE)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

_ = ConfusionMatrixDisplay.from_estimator(bundled_pipeline, X_test, y_test)
```
